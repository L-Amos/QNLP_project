# Plotter Notebook
This notebook seamlessly plots data gathered from training runs done on the GPU.

```python
import glob
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

### File Parsing Function

```python
def readfiles(files):
    filedumps = [np.genfromtxt(filename, delimiter=',') for filename in files]
    return filedumps
```

## Plots During Testing

### Plot Specific Data

```python
# Train Costs
TRAIN_FILES = ["num_train_val_pairs/train_costs_500-100.csv"]
VAL_FILES = ["num_train_val_pairs/val_costs_500-100.csv"]
# Val Costs
train_costs = readfiles(TRAIN_FILES)
val_costs = readfiles(VAL_FILES)
# Plot
plt.plot(train_costs[0], color='darkred', linestyle='-', label="Train Costs: 500 train, 100 val")
plt.plot(val_costs[0], color='darkred', linestyle='--', label="Val Costs: 500 train, 100 val")
# plt.plot(train_costs[1], color='darkblue', linestyle='-', label="Train Costs: 1000 train, 200 val")
# plt.plot(val_costs[1], color='darkblue', linestyle='--', label="Val Costs: 1000 train, 200 val")

plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.title(f"Comparison of Model Convergences")
plt.legend()
plt.grid()
plt.show()
```

```python
# Train Costs
TRAIN_FILES = ["num_train_val_pairs/train_corr_500-100.csv"]
VAL_FILES = ["num_train_val_pairs/val_corr_500-100.csv"]
# Val Costs
train_costs = readfiles(TRAIN_FILES)
val_costs = readfiles(VAL_FILES)
# Plot
plt.plot(train_costs[0], color='darkred', linestyle='-', label="Train Costs: 500 train, 100 val")
plt.plot(val_costs[0], color='darkred', linestyle='--', label="Val Costs: 500 train, 100 val")
# plt.plot(train_costs[1], color='darkblue', linestyle='-', label="Train Costs: 1000 train, 200 val")
# plt.plot(val_costs[1], color='darkblue', linestyle='--', label="Val Costs: 1000 train, 200 val")

plt.xlabel("Epoch")
plt.ylabel("Correlation Coefficient")
plt.title(f"Comparison of Model Correlations")
plt.legend()
plt.grid()
plt.show()
```
